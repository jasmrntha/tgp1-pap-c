## Pembersihan Data (Data Cleaning)

Tujuan: Memperbaiki atau menangani masalah dalam data, seperti missing values, duplikasi, dan outlier.

### Penanganan Missing Value

Pada visualisasi data distribusi usia korban, terdapat banyak sekali usia berusia 0,
Usia 0 ini kami anggap sebagai missing value

```python
# Handling vict age = 0 value
median_age = data['Vict Age'][data['Vict Age'] != 0].median()
data['Vict Age'] = data['Vict Age'].replace(0, median_age)

# Missing value usia korban (NaN)
data['Vict Age'].fillna(median_age)
print("number of age is null: ",data['Vict Age'].isnull().sum())
```

```python
# Missing value jenis kelamin korban
data['Vict Sex'] = data['Vict Sex'].fillna(data['Vict Sex'].mode()[0])
print("number of sex is null: ",data['Vict Sex'].isnull().sum())

# Missing value ras korban
data['Vict Descent'] = data['Vict Descent'].fillna(data['Vict Descent'].mode()[0])
print("number of descent is null: ",data['Vict Descent'].isnull().sum())
```

### Menghapus Data yang Tidak Relevan

```python
# Menghapus kolom 'DR_NO', 'Rpt Dist No', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4' karena tidak relevan
data = data.drop(columns=['DR_NO', 'Rpt Dist No', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4'])
data.info()
```

### Menangani Duplikasi

```python
data = data.drop_duplicates()
data.info()
```

### Penanganan Outlier

```python
# Data Vict Age sebelum penanganan outlier
plt.figure(figsize=(8,6))
sns.boxplot(x=data['Vict Age'])
plt.title('Boxplot Victim Age')
plt.xlabel('Victim Age')
plt.show()
```

```python
# Penanganan menggunakan interquartile 
Q1 = data['Vict Age'].quantile(0.25)
Q3 = data['Vict Age'].quantile(0.75)
IQR = Q3 - Q1

# Filtering out Outliers
data = data[~((data['Vict Age'] < (Q1 - 1.5 * IQR)) | (data['Vict Age'] > (Q3 + 1.5 * IQR)))]

# Vict Age setelah penanganan outlier
plt.figure(figsize=(8,6))
sns.boxplot(x=data['Vict Age'])
plt.title('Boxplot Victim Age')
plt.xlabel('Victim Age')
plt.show()
```

## Transformasi Data (Data Transformation)

Tujuan: Menyiapkan dataset agar siap untuk digunakan dalam pemodelan prediktif dengan melakukan transformasi fitur yang relevan.

```python
# Membuat variabel baru untuk data yang akan ditransformasi
data_tform = data
data_tform
```

### Konversi Waktu dan Tanggal

```python
# Date occ awalnya bertipe data str object
# Time occ awalnya bertipe data int

# Mengubah Date occ menjadi datetime
data_tform['DATE OCC'] = pd.to_datetime(data_tform['DATE OCC'])

# Mengubah Time occ menjadi string
data_tform['TIME OCC_str'] = data_tform['TIME OCC'].apply(lambda x: f'{x:04d}'[:2] + ':' + f'{x:04d}'[2:])

# Menggabungkan Date dan Time 
data_tform['DATETIME OCC'] = pd.to_datetime(data_tform['DATE OCC'].dt.strftime('%Y-%m-%d') + ' ' + data_tform['TIME OCC_str'])

# Hari dalam seminggu
data_tform['day_of_week'] = data_tform['DATETIME OCC'].dt.day_name()

# Bulan
data_tform['month'] = data_tform['DATETIME OCC'].dt.month_name()
```

```python
# drop column date occ dan time occ
data_tform = data_tform.drop(columns=['DATE OCC','TIME OCC'])
data_tform.head()
```

fungsi 'data_tform['TIME OCC'].apply(lambda x: f'{x:04d}'[:2] + ':' + f'{x:04d}'[2:])'

adalah sebuah fungsi untuk mengubah Time occ yang awalnya 4 digit integer menjadi string

.apply() function digunakan untuk mengaplikasikan fungsi kesetiap element dalam TIME OCC

lambda x: merupakan syntax yang mengambil argument x (yaitu TIME OCC)

expression f'{x:04d}' merupakan f-string syntax
x:04d berarti mengubah format integer menjadi 4 karakter string 

kemudian f'{x:04d}'[:2] mengambil 2 karakter pertama yang merepresentasikan jam

dan f'{x:04d}'[2:] mengambil 2 karakter terakhir yang merepresentasikan menit.


### One Hot Encoding

```python
# One-Hot Encoding Vict Sex
one_hot = pd.get_dummies(data_tform['Vict Sex'])
one_hot.head()
```

```python
# Dalam deskripsi data hanya terdapat (F - Female M - Male X - Unknown)
# Tetapi kolom terdapat 2 kolom yang tidak relevan yaitu '-' dan 'H'

# Replace '-' with 'X' unknown
data_tform['Vict Sex'] = data_tform['Vict Sex'].replace('-','X')

# Replace 'H' with 'X' Unknown
data_tform['Vict Sex'] = data_tform['Vict Sex'].replace('H', 'X')
```

```python
# Mengulang one hot encoding dan mengubah nama kolom
one_hot = pd.get_dummies(data_tform['Vict Sex'])
one_hot= one_hot.rename(columns={'F': 'Female', 'M': 'Male', 'X': 'Unknown gender'})
one_hot.head()
```

```python
# Menggabungkan kolom Female, Male, Unknown gender dan membuang kolom Vict Sex
data_tform = data_tform.drop(columns='Vict Sex')
data_tform = data_tform.join(one_hot)
data_tform.head()
```

```python
# One-Hot Encoding Satatus Desc
one_hot_status = pd.get_dummies(data_tform['Status Desc'])
one_hot_status.head(10)
```

```python
# Remove Status Desc dan join one hot encoded ke data
data_tform = data_tform.drop(columns='Status Desc')
data_tform = data_tform.join(one_hot_status)

data_tform.head(10)
```

### Korelasi Antar Fitur

```python
# Korelasi Antar Fitur
numerical_features = ['Vict Age', 'LAT', 'LON']
correlation_matrix = data_for_new_feature[numerical_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, vmin=-1, vmax=1, linewidths=0.5, square=True)
plt.title('Correlation Matrix Antar Fitur Numerik')
plt.show()
```

Diagram diatas menjelaskan korelasi antar fitur numerik, yaitu: **Vict Age**, **LAT**, **LON**.
- Apabila nilai semakin mendekati 1 maka kedua fitur tersebut saling memiliki korelasi satu sama lain, pada diagram diatas ditunjukkan dengan korelasi fitur yang sama itu selalu 1.
- Antar fitur **LAT** dan **LON** nilai korelasinya adalah -1 yang berarti ketika satu variabel meningkat, yang lain menurun dengan cara yang benar-benar linier.
- Antar fitur **Vict Age** dan **LAT** korelasinya adalah -0.0004, menunjukkan korelasi negatif yang sangat lemah. Ini berarti hampir tidak ada hubungan linier antara kedua variabel ini.
- Antar fitur **Vict Age** dan **LON** korelasinya adalah 0.0017, menunjukkan korelasi positif yang sangat lemah, yang juga berarti hampir tidak ada hubungan linier.

## Evaluasi dan Kesimpulan

Tujuan: Mengumpulkan hasil eksplorasi dan prapemrosesan data, serta menyiapkan dataset untuk pemodelan prediktif.

#### Pola Utama

**a. Distribusi Usia Korban:**

Usia korban kejahatan memiliki distribusi yang bervariasi, dengan median usia diisi untuk menggantikan nilai usia 0, yang dianggap sebagai missing value.
Setelah pembersihan, boxplot menunjukkan bahwa outlier telah ditangani dengan baik.

**b. Waktu dan Hari Terjadinya Kejahatan:**

Distribusi kejahatan menunjukkan pola harian dan bulanan. Misalnya, terdapat perbedaan signifikan dalam frekuensi kejahatan berdasarkan jam, dengan puncaknya terjadi pada waktu malam (setelah pukul 18.00).
Frekuensi kejahatan juga bervariasi sepanjang minggu, dengan kemungkinan lebih tinggi di hari-hari tertentu.

**c. Geografi Kejahatan:**

Pemetaan kejahatan berdasarkan lokasi menunjukkan konsentrasi tinggi kejahatan di lokasi tertentu, yang dapat membantu dalam penentuan area rawan kejahatan.
Analisis lebih lanjut terhadap area menunjukkan bahwa beberapa daerah lebih rentan terhadap kejahatan.

**d. Korelasi Antar Fitur:**

Terdapat korelasi yang sangat lemah antara usia korban dan koordinat geografis (LAT dan LON). Ini menunjukkan bahwa lokasi dan demografi mungkin tidak berhubungan secara langsung dengan usia korban.
Namun, hubungan yang negatif antara LAT dan LON menunjukkan bahwa peningkatan satu variabel diikuti oleh penurunan yang lain.

#### Fitur yang Paling Relevan untuk Memprediksi Jenis Kejahatan

**a. Vict Age:** Usia korban dapat memberikan wawasan tentang kelompok demografis yang lebih rentan terhadap jenis kejahatan tertentu.

**b. Waktu Kejadian:** Waktu (jam dan apakah itu siang atau malam) adalah fitur kunci, karena beberapa jenis kejahatan lebih mungkin terjadi pada waktu tertentu.

**c. Hari dalam Minggu:** Frekuensi kejahatan yang berbeda pada hari-hari tertentu memberikan informasi tambahan.

**d. Geographic Zone:** Zona geografis dapat membantu dalam memahami di mana kejahatan lebih sering terjadi dan mungkin dipengaruhi oleh faktor sosial atau ekonomi di daerah tersebut.

**e. Kelamin Korban:** Fitur ini dapat menjadi indikator penting dalam menentukan jenis kejahatan tertentu yang mungkin lebih sering menargetkan gender tertentu.

#### Dataset Akhir
Setelah melakukan pembersihan dan transformasi data, dataset akhir siap digunakan untuk pemodelan prediktif. Berikut adalah langkah-langkah yang dilakukan:

1. **Pembersihan Missing Values:** Nilai usia 0 diubah menjadi median usia, dan nilai yang hilang pada kolom lainnya diisi dengan mode.

2. **Penghapusan Kolom yang Tidak Relevan:** Kolom yang tidak mendukung analisis dihapus.

3. **Penanganan Duplikasi:** Data duplikat dihapus untuk memastikan setiap entri unik.

4. **Penanganan Outlier:** Outlier pada kolom usia korban ditangani menggunakan metode IQR.

5. **Transformasi Tanggal dan Waktu:** Tanggal dan waktu diubah menjadi format yang sesuai untuk analisis lebih lanjut.

6. **One-Hot Encoding:** Variabel kategori diubah menjadi format numerik untuk digunakan dalam pemodelan.

7. **Standarisasi:** Usia korban distandarisasi menggunakan Z-score untuk normalisasi data.

8. **Penambahan fitur baru:** Membuat fitur baru dengan tipe biner untuk waktu kejahatan dan zona geografis berdasarkan area kejahatan
