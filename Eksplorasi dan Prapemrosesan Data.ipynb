## Pembersihan Data (Data Cleaning)

Tujuan: Memperbaiki atau menangani masalah dalam data, seperti missing values, duplikasi, dan outlier.

```python
# Penanganan Missing Values
data['Vict Age'].fillna(data['Vict Age'].median(), inplace=True)
```

```python
# Menghapus Data yang Tidak Relevan
data.drop(columns=['DR_NO', 'Rpt Dist No', 'Crm Cd 1'], inplace=True)
```

```python
# Menangani Duplikasi
data.drop_duplicates(inplace=True)
```

```python
# Penanganan Outlier
Q1 = data['Vict Age'].quantile(0.25)
Q3 = data['Vict Age'].quantile(0.75)
IQR = Q3 - Q1

# Filtering out Outliers
data_clean = data[~((data['Vict Age'] < (Q1 - 1.5 * IQR)) | (data['Vict Age'] > (Q3 + 1.5 * IQR)))]
```
